```python
import pandas as pd
import pandas_ta as ta

data = pd.read_csv(f"../data/us100/NAS100_M1_202101040100_202212232354.csv", sep='\t')
data['<DATE>'] = data['<DATE>'] + ' ' + data['<TIME>']
data.rename(columns={
    "<DATE>": "date",
    "<OPEN>": "open",
    "<HIGH>": "high",
    "<LOW>": "low",
    "<CLOSE>": "close",
    "<SPREAD>": "spread"
}, inplace=True)
data = data[['date', 'open', 'high', 'low', 'close', 'spread']]
data.reset_index(drop=True, inplace=True)
data['date'] = pd.to_datetime(data['date'], format='%Y.%m.%d')
data.loc[data["spread"] == 0, "spread"] = 1
```

```python
data.ta.sma(length=20, append=True)
data.ta.sma(length=50, append=True)
data.dropna(inplace=True)
data.reset_index(drop=True, inplace=True)
```

```python
_test = pd.DataFrame(data={
    "A": [1, 2, 4, 5, 6, 5, 4, 2, 1],
    "B": [3] * 9
})
_test["up"] = _test["A"] > _test["B"]
_test["down"] = _test["A"] < _test["B"]
_test["down"] = _test["down"].shift(1)
_test
```

```python
_test["crossover"] = _test["up"] & _test["down"]
_test.drop(["up", "down"], axis=1, inplace=True)
_test
```
